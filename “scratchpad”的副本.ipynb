```
import pandas as pd
import numpy as np

def calculate_month_macd(daily_data, short_period=12, long_period=26, signal_period=9):
    # 将每日交易数据按月聚合
    monthly_data = daily_data.resample('M').last()

    # 计算收盘价的短期和长期指数移动平均线（EMA）
    monthly_data['month_EMA_short'] = monthly_data['Closing'].ewm(span=short_period, adjust=False).mean()
    monthly_data['month_EMA_long'] = monthly_data['Closing'].ewm(span=long_period, adjust=False).mean()

    # 计算MACD指标线
    monthly_data['month_DIFF_line'] = monthly_data['month_EMA_short'] - monthly_data['month_EMA_long']

    # 计算信号线（MACD的短期指数移动平均线）
    monthly_data['month_DEA_line'] = monthly_data['month_DIFF_line'].ewm(span=signal_period, adjust=False).mean()

    # 计算MACD柱状图（MACD线与信号线的差值）
    monthly_data['month_MACD_histogram'] = 2*(monthly_data['month_DIFF_line'] - monthly_data['month_DEA_line'])

    return monthly_data

def calculate_week_macd(daily_data, short_period=12, long_period=26, signal_period=9):
    # 将每日交易数据按月聚合
    monthly_data = daily_data.resample('W').last()

    # 计算收盘价的短期和长期指数移动平均线（EMA）
    monthly_data['week_EMA_short'] = monthly_data['Closing'].ewm(span=short_period, adjust=False).mean()
    monthly_data['week_EMA_long'] = monthly_data['Closing'].ewm(span=long_period, adjust=False).mean()

    # 计算MACD指标线
    monthly_data['week_DIFF_line'] = monthly_data['week_EMA_short'] - monthly_data['week_EMA_long']

    # 计算信号线（MACD的短期指数移动平均线）
    monthly_data['week_DEA_line'] = monthly_data['week_DIFF_line'].ewm(span=signal_period, adjust=False).mean()

    # 计算MACD柱状图（MACD线与信号线的差值）
    monthly_data['week_MACD_histogram'] = 2*(monthly_data['week_DIFF_line'] - monthly_data['week_DEA_line'])

    return monthly_data



# # 示例每日交易数据
# daily_data = pd.DataFrame({
#     'Date': pd.date_range('2022-01-01', periods=100, freq='D'),
#     'Closing': np.random.randint(100, 200, size=100)
# })
# daily_data.set_index('Date', inplace=True)

# # 计算当天的月线级别MACD值，每日的值
# monthly_macd = calculate_daily_macd(daily_data)

# 打印结果
# print(monthly_macd)
```
